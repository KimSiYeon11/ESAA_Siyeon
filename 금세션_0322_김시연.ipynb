### Step 1. Import the necessary libraries:

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

```python
df=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv", sep=',')
df.head()
```

### Step 3. Assign it to a variable called tips

```python
tips=df
```

### Step 4. Delete the Unnamed 0 column

```python
tips= tips.drop(labels='Unnamed: 0', axis=1)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(8,6))
sns.histplot(data=tips,x='total_bill')
plt.title('Histogram of total_bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter plot of total_bill vs tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
#   total_bill, tip, size   
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='total_bill', y='tip', size='size', sizes=(20, 200))
plt.title('Bubble Chart of total_bill, tip, and size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
tips['day'].unique()
```

```python
#    total_bill  
mean_total_bill_by_day= tips.groupby('day')['total_bill'].mean().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=mean_total_bill_by_day, x='day', y='total_bill', palette='Set2')
plt.title('Mean Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Mean Total Bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=tips, x='tip', y='day', hue='sex', palette='Set2')
plt.title('Scatter plot of tip, day, sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time',palette='Set3')
plt.title('Box Plot of Total Bill per Day by Time')
plt.xlabel('day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(12, 6))

# Lunch    
plt.subplot(1,2,1)
sns.histplot(tips[tips['time']=='Lunch']['tip'], bins=15)
plt.title('Histogram of Lunch')
plt.xlabel('Tip')
plt.ylabel('Frequency')

# Dinner    
plt.subplot(1,2,2)
sns.histplot(tips[tips['time']=='Dinner']['tip'], bins=15)
plt.title('Histogram of Dinner')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(12,6))

#    
plt.subplot(1,2,1)
sns.scatterplot(data=tips[tips['sex']=='Male'], x='total_bill', y='tip', hue='smoker')
plt.title('Scatter plot of Tip vs Total Bill for Male')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Smoker')

#    
plt.subplot(1,2,2)
sns.scatterplot(data=tips[tips['sex']=='Female'], x='total_bill', y='tip', hue='smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Smoker')

plt.show()
```
