### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

```python
data=pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep='|')
data.head()
```

### Step 3. Assign it to a variable called users.

```python
print(data.columns)
data['user_id']
```

### Step 4. Discover what is the mean age per occupation

```python
data.groupby("occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
data['is_male']=data.gender.apply(lambda x : True if x=='M' else False)
data.is_male
```

```python
data.columns
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
print(data.groupby('occupation').age.min())
print(data.groupby('occupation').age.max())
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
data.groupby(['occupation','gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_ocup = data.groupby(['occupation', 'gender']).agg({'gender':'count'})
occup_count = data.groupby(['occupation']).count()
occup_gender = gender_ocup.div(occup_count, level = 'occupation') # level : 그룹화의 기준을 나타내는 파라미
occup_gender.loc[:,'gender']
```
