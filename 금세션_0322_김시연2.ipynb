### Step 1. Import the necessary libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Step 2. Create the DataFrame that should look like the one below.
*참고*

```
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

````python
df=pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]})
df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='preTestScore', y='postTestScore', size='age', sizes=(20,300))
plt.title('Bubble Chart of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
# 성별에 따라 데이터 분할
male=df[df['female']==0]
female=df[df['female']==1]

# 산점도 그리기
plt.figure(figsize=(8,6))
sns.scatterplot(data=male, x='preTestScore', y='postTestScore', color='blue')
sns.scatterplot(data=female, x='preTestScore', y='postTestScore', color='red')
plt.title('Scatter Plot of preTestScore vs postTestScore by Sex')
plt.xlabel('PreTest Score')
plt.ylabel('PostTest Score')
plt.show()
````
